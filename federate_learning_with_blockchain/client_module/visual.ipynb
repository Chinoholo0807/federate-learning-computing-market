```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_plot(x,y,xlabel='round',ylabel='acc',learning_rate=0.01,
                n_client=10,n_participators=5,epochs=5,batch_size=10,
                model_name='mnist_2nn',aggreagate_method='fed_avg',attacker_prop=0.0):
    plt.plot(x,y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    title = 'federate learning:n_client({}) n_participators({}) epochs({}) lr({})batch_size({}) model_name({}) aggregate({}) attaker({}%)'.format(
        n_client,
        n_participators,
        epochs,
        learning_rate,
        batch_size,
        model_name,
        aggreagate_method,
        attacker_prop
    )
    plt.title(title)
    plt.show()
```

```python
PATH = r'.\reports\results\2022-03-07_15-52-16@npart5_lr0.01_bs10_ep5_mnist_2nn_fed_avg.csv'
df = pd.read_csv(PATH)
df.head()
```

```python
draw_plot(df['round'],df['global_acc'],ylabel='global_acc')
draw_plot(df['round'],df['global_loss'],ylabel='global_loss')
```
