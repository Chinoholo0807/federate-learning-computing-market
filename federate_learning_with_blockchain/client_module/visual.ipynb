```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_plot(x,y,xlabel='round',ylabel='acc',aggreagate_method='fed_avg',attacker_prop=0.0):
    plt.plot(x,y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(f'federate learning :aggregate({aggreagate_method}),attacker_prop({attacker_prop})')
    plt.show()
```

```python
PATH = r'.\reports\results\2022-03-06_17-15-06@npart4_lr0.01_bs10_ep4_mnist_2nn.csv'
df = pd.read_csv(PATH)
df.head()
```

```python
draw_plot(df['round'],df['global_acc'],ylabel='global_acc')
draw_plot(df['round'],df['global_loss'],ylabel='global_loss')
```
