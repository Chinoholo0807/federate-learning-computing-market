```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_plot(x,y,aggregate,attacker):
    plt.plot(x,y)
    plt.xlabel('round')
    plt.ylabel('acc')
    plt.title(f'federate learning aggregate({aggregate}) attaker({attacker})')
    plt.show()
```

```python
PATH = r'.\results\2022-03-06_13-22-57_result.txt'
df = pd.read_csv(PATH,names=['round','accuracy'],header=None)
draw_plot(df['round'],df['accuracy'],0,10000)
```

```python
PATH = r'.\results\2022-03-06_13-46-06_result.txt'
df = pd.read_csv(PATH,names=['round','accuracy'],header=None)
draw_plot(df['round'],df['accuracy'],1,10000)
```
